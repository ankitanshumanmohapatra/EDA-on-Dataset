```python
import pandas as pd
import numpy as np
```

```python
#MATPLOTLIB
import matplotlib.pyplot as plt
%matplotlib inline
#seaborn
import seaborn as sns
```

```python
plt.scatter(df.index,df['SepalWidthCm'])
plt.show()
```

```python
sns.scatterplot(x=df.index,y=df['SepalWidthCm'],hue=df['Species'])
```

```python
plt.figure(figsize=(6,6))
plt.title('Line plot of petal length')
plt.xlabel('index',fontsize=20)
plt.ylabel('petal length',fontsize=20)
plt.plot(df.index,df['PetalLengthCm'],markevery=1,marker='d')
for name,group in df.groupby('Species'):
  plt.plot(group.index,group['PetalLengthCm'], label=name,markevery=1,marker='d')
plt.legend()
plt.show()
```

```python
#strip plot
sns.stripplot(y=df['SepalWidthCm'])
```

```python
sns.stripplot(x=df['Species'],y=df['SepalWidthCm'])
```

```python
#Swarm plot
sns.set(rc={'figure.figsize':(5,5)})
sns.swarmplot(x=df['SepalWidthCm'])
```

```python
plt.hist(df['PetalWidthCm'])
```

```python
plt.figure(figsize=(5,5))
df['PetalLengthCm'].plot(kind='density')
```
